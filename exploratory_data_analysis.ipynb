```python
# top 10 countries

data["Country"].value_counts().head(10).plot(kind='pie',autopct = '%1.0f%%')
```

The exploratory data analysis of this dataset focus on the 'Country' attribute and 'ComputerSkills' attribute.
The first plot is a pie chart distribution among the top ten countries. The United States of America has the largest representation with 34% of the entries. Germany and India are equally represented, each with 12%. Canada follows with 11%, while France and Brazil each have 6%. Spain, the Netherlands, and Poland round off the chart with 4% each. The chart indicates a strong concentration in the USA, with a significant drop-off to other countries, which are more evenly distributed in terms of representation.


```python
# % of people with 10+ computerskills
data["ComputerSkillsLevel"] = pd.cut(x = data["ComputerSkills"], 
                                 bins = [0, 10, 20, 30, 100],
                                   labels = ['<10', '10-20', '20-30', '30+'])
print("ComputerSkillsLevel: ")
print(data["ComputerSkillsLevel"].value_counts())
```

```python
data["ComputerSkillsLevel"].value_counts().plot(kind='pie',autopct = '%1.0f%%')
```

The second plot is a pie chart representing the computer skills level of individuals within the dataset, categorized into four groups based on the number of skills: less than 10, between 10 and 20, between 20 and 30, and more than 30 skills. The largest segment, comprising 48% of the dataset, has a skill range of 10 to 20. The second-largest group, representing 37%, possesses fewer than 10 skills.
Those with skills ranging from 20 to 30 make up 12%, while individuals with more than 30 skills are the smallest group, accounting for just 2% of the dataset. This distribution indicates that the majority of individuals have a moderate level of computer skills, with very few reaching the high skill level of 30+.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# For example, to plot a histogram of a numeric column:
sns.histplot(data['YearsCode'])
plt.show()
```

```python
# Loop through each column in the dataframe
df = data
for column in df.columns:
    # If the column is numeric, plot a histogram
    if pd.api.types.is_numeric_dtype(df[column]):
        plt.figure(figsize=(10, 6))
        sns.histplot(df[column], kde=False, bins=30)
        plt.title(f'Histogram for {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()
    # If the column is categorical, plot a countplot
    elif pd.api.types.is_categorical_dtype(df[column]) or df[column].dtype == 'O':
        plt.figure(figsize=(10, 6))
        sns.countplot(y=df[column], order=df[column].value_counts().index)
        plt.title(f'Count plot for {column}')
        plt.xlabel('Count')
        plt.ylabel(column)
        plt.show()
```
